```python
# --- 3. Broadcasting example ---
x = np.array([[1],[2],[3]])       # shape (3, 1)
y = np.array([10, 20, 30])        # shape (3,)
print(x + y)                      # broadcasts to (3, 3)
```

```python
# --- 4. Vectorized ops (no loops!) ---
def dot_product(a, b):
    return np.dot(a, b)

def normalize(v):
    return v / np.linalg.norm(v)

def matrix_multiply(A, B):
    return A @ B                  # or np.matmul(A, B)
```

```python
# --- 5. Test your functions ---
v = np.array([3.0, 4.0])
print("Normalized:", normalize(v))  # [0.6, 0.8]

A = np.random.randn(2, 3)
B = np.random.randn(3, 2)
print("MatMul shape:", matrix_multiply(A, B).shape)  # (2, 2)
```
