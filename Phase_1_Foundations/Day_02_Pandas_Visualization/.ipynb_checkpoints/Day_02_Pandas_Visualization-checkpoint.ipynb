```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#-- Load dataset---
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
df['species'] = df['target'].map({0: 'setosa', 1:'versicolor', 2:'virgincia'})

```

```python

#--Explore the data---
print(df.shape)
print(df.head())
print(df.describe())
print(df.info())

#--Check for missing values--
print(df.isnull().sum())
```

```python
#--Filter & select--
setosa= df[df['species']== 'setosa']
print("setosa rows:", len(setosa))

#--Add a new feature (petal area)
df['petal_area'] = df['petal length (cm)'] * df['petal width (cm)']
```

```python
#-- Matplotlib plots--
fig, axes = plt.subplots(1,2, figsize=(12,4))

axes[0].hist(df['sepal length (cm)'], bins = 20, color='steelblue', edgecolor='white')
axes[0].set_title('Sepal Length Distribution')

axes[1].scatter(df['sepal length (cm)'], df['sepal width (cm)'],
                c=df['target'], cmap='viridis', alpha=0.7)

axes[1].set_title('Sepal Length vs Width')
plt.tight_layout()
plt.show()
```

```python

#--Seaborn - the 3 key ML plots--

#--Class balance,3))
plt.figure(figsize=(5,3))
sns.countplot(x='species', data=df, hue='species', legend=False, palette='Set2')
plt.title('Class_Balance')
plt.show()

# Feature correlation heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(df.drop(columns=['target','species']).corr(),
            annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature Correlations')
plt.show()

# Pairplot — see all feature relationships at once
sns.pairplot(df, hue='species', vars=iris.feature_names, plot_kws={'alpha':0.6})
plt.show()

```
